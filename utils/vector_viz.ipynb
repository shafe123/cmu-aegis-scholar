# Vector Averaging Visualization

This notebook demonstrates how to visualize multiple vectors and their average vector in both 2D and 3D space.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.decomposition import PCA

# Set style for better-looking plots
plt.style.use('seaborn-v0_8-darkgrid')
```

## 2D Vector Visualization

Let's create several 2D vectors and visualize their average.

```python
# Create several 2D vectors
vectors_2d = np.array([
    [3, 2],
    [1, 4],
    [4, 1],
    [2, 3],
    [3.5, 2.5]
])

# Calculate the average vector
avg_vector_2d = np.mean(vectors_2d, axis=0)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 8))

# Plot individual vectors
for i, vec in enumerate(vectors_2d):
    ax.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, 
              color='steelblue', alpha=0.6, width=0.006, label='Individual Vectors' if i == 0 else '')

# Plot average vector with thicker line
ax.quiver(0, 0, avg_vector_2d[0], avg_vector_2d[1], angles='xy', scale_units='xy', scale=1,
          color='red', width=0.012, label='Average Vector', linewidth=2)

# Set plot properties
ax.set_xlim(-0.5, 5)
ax.set_ylim(-0.5, 5)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_title('2D Vectors and Their Average', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='upper left')
ax.set_aspect('equal')

# Add text showing the average vector values
ax.text(avg_vector_2d[0] + 0.2, avg_vector_2d[1] + 0.2, 
        f'Avg: ({avg_vector_2d[0]:.2f}, {avg_vector_2d[1]:.2f})',
        fontsize=10, color='red', fontweight='bold')

plt.tight_layout()
plt.show()

print(f"Individual vectors:\n{vectors_2d}")
print(f"\nAverage vector: [{avg_vector_2d[0]:.2f}, {avg_vector_2d[1]:.2f}]")
```

## 3D Vector Visualization

Now let's visualize vectors in 3D space.

```python
# Create several 3D vectors
vectors_3d = np.array([
    [2, 3, 1],
    [1, 2, 4],
    [3, 1, 2],
    [2, 2, 3],
    [1.5, 2.5, 2],
    [2.5, 1.5, 2.5]
])

# Calculate the average vector
avg_vector_3d = np.mean(vectors_3d, axis=0)

# Create 3D plot
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot individual vectors
for i, vec in enumerate(vectors_3d):
    ax.quiver(0, 0, 0, vec[0], vec[1], vec[2], 
              color='steelblue', alpha=0.6, arrow_length_ratio=0.15,
              linewidth=2, label='Individual Vectors' if i == 0 else '')

# Plot average vector with thicker line
ax.quiver(0, 0, 0, avg_vector_3d[0], avg_vector_3d[1], avg_vector_3d[2],
          color='red', arrow_length_ratio=0.15, linewidth=4,
          label='Average Vector')

# Set plot properties
ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
ax.set_zlim([0, 5])
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)
ax.set_title('3D Vectors and Their Average', fontsize=14, fontweight='bold')
ax.legend(loc='upper left')

# Add text showing the average vector values
ax.text(avg_vector_3d[0], avg_vector_3d[1], avg_vector_3d[2] + 0.3,
        f'Avg: ({avg_vector_3d[0]:.2f}, {avg_vector_3d[1]:.2f}, {avg_vector_3d[2]:.2f})',
        fontsize=10, color='red', fontweight='bold')

plt.tight_layout()
plt.show()

print(f"Individual vectors:\n{vectors_3d}")
print(f"\nAverage vector: [{avg_vector_3d[0]:.2f}, {avg_vector_3d[1]:.2f}, {avg_vector_3d[2]:.2f}]")
```

## Multiple Groups of Vectors

Let's visualize multiple groups of vectors, each with their own average.

```python
# Create two groups of 2D vectors
group1_vectors = np.array([
    [1, 3],
    [1.5, 3.5],
    [0.5, 2.5],
    [1.2, 3.2]
])

group2_vectors = np.array([
    [4, 1],
    [4.5, 1.5],
    [3.5, 0.5],
    [4.2, 1.2]
])

# Calculate averages for each group
avg_group1 = np.mean(group1_vectors, axis=0)
avg_group2 = np.mean(group2_vectors, axis=0)

# Create the plot
fig, ax = plt.subplots(figsize=(12, 10))

# Plot Group 1 vectors
for i, vec in enumerate(group1_vectors):
    ax.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1,
              color='blue', alpha=0.4, width=0.005)

# Plot Group 2 vectors
for i, vec in enumerate(group2_vectors):
    ax.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1,
              color='green', alpha=0.4, width=0.005)

# Plot group averages
ax.quiver(0, 0, avg_group1[0], avg_group1[1], angles='xy', scale_units='xy', scale=1,
          color='darkblue', width=0.01, label='Group 1 Average')
ax.plot([], [], color='blue', alpha=0.4, linewidth=3, label='Group 1 Vectors')

ax.quiver(0, 0, avg_group2[0], avg_group2[1], angles='xy', scale_units='xy', scale=1,
          color='darkgreen', width=0.01, label='Group 2 Average')
ax.plot([], [], color='green', alpha=0.4, linewidth=3, label='Group 2 Vectors')

# Dummy plots for legend

# Set plot properties
ax.set_xlim(-0.5, 5.5)
ax.set_ylim(-0.5, 4.5)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_title('Multiple Groups of Vectors with Individual and Overall Averages', 
             fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='upper right', fontsize=10)
ax.set_title('Multiple Groups of Vectors with Group Averages')

# Add text annotations
ax.text(avg_group1[0] + 0.15, avg_group1[1] + 0.15,
        f'G1: ({avg_group1[0]:.2f}, {avg_group1[1]:.2f})',
        fontsize=9, color='darkblue', fontweight='bold')
ax.text(avg_group2[0] + 0.15, avg_group2[1] - 0.3,
        f'G2: ({avg_group2[0]:.2f}, {avg_group2[1]:.2f})',
        fontsize=9, color='darkgreen', fontweight='bold')


plt.tight_layout()
plt.show()
print(f"Group 1 average: [{avg_group1[0]:.2f}, {avg_group1[1]:.2f}]")
print(f"Group 2 average: [{avg_group2[0]:.2f}, {avg_group2[1]:.2f}]")
```

## Multiple Groups of Vectors in 3D

Now let's visualize the same concept in 3D space.

```python
# Create two groups of 3D vectors
group1_vectors_3d = np.array([
    [1, 3, 2],
    [1.5, 3.5, 2.5],
    [0.5, 2.5, 1.5],
    [1.2, 3.2, 2.2],
    [0.8, 2.8, 1.8]
])

group2_vectors_3d = np.array([
    [4, 1, 2],
    [4.5, 1.5, 2.5],
    [3.5, 0.5, 1.5],
    [4.2, 1.2, 2.2],
    [3.8, 0.8, 1.8]
])

# Calculate averages for each group
avg_group1_3d = np.mean(group1_vectors_3d, axis=0)
avg_group2_3d = np.mean(group2_vectors_3d, axis=0)

# Create an "incoming query" vector closer to group 1
incoming_query = np.array([1.8, 2.5, 1.8])

# Create 3D plot
fig = plt.figure(figsize=(14, 11))
ax = fig.add_subplot(111, projection='3d')

# Plot Group 1 vectors
for i, vec in enumerate(group1_vectors_3d):
    ax.quiver(0, 0, 0, vec[0], vec[1], vec[2],
              color='blue', alpha=0.4, arrow_length_ratio=0.15,
              linewidth=2)
ax.quiver(0, 0, 0, avg_group1_3d[0], avg_group1_3d[1], avg_group1_3d[2],
          color='darkblue', arrow_length_ratio=0.15, linewidth=4,
          label='Group 1 Average')
ax.plot([], [], color='blue', alpha=0.4, linewidth=3, label='Group 1 Vectors')

# Plot Group 2 vectors
for i, vec in enumerate(group2_vectors_3d):
    ax.quiver(0, 0, 0, vec[0], vec[1], vec[2],
              color='green', alpha=0.4, arrow_length_ratio=0.15,
              linewidth=2)
ax.quiver(0, 0, 0, avg_group2_3d[0], avg_group2_3d[1], avg_group2_3d[2],
          color='darkgreen', arrow_length_ratio=0.15, linewidth=4,
          label='Group 2 Average')
ax.plot([], [], color='green', alpha=0.4, linewidth=3, label='Group 2 Vectors')

# Plot incoming query vector
ax.quiver(0, 0, 0, incoming_query[0], incoming_query[1], incoming_query[2],
          color='magenta', arrow_length_ratio=0.15, linewidth=4,
          label='Incoming Query')

# Set plot properties
ax.set_xlim([0, 5])
ax.set_ylim([0, 4])
ax.set_zlim([0, 3])
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)
ax.set_title('Multiple Groups of Vectors in 3D with Group Averages', 
             fontsize=14, fontweight='bold')
ax.legend(loc='upper left', fontsize=10)

# Add text annotations
ax.text(avg_group1_3d[0], avg_group1_3d[1], avg_group1_3d[2] + 0.2,
        f'G1: ({avg_group1_3d[0]:.2f}, {avg_group1_3d[1]:.2f}, {avg_group1_3d[2]:.2f})',
        fontsize=9, color='darkblue', fontweight='bold')
ax.text(avg_group2_3d[0], avg_group2_3d[1], avg_group2_3d[2] + 0.2,
        f'G2: ({avg_group2_3d[0]:.2f}, {avg_group2_3d[1]:.2f}, {avg_group2_3d[2]:.2f})',
        fontsize=9, color='darkgreen', fontweight='bold')
ax.text(incoming_query[0], incoming_query[1], incoming_query[2] + 0.2,
        f'Query: ({incoming_query[0]:.2f}, {incoming_query[1]:.2f}, {incoming_query[2]:.2f})',
        fontsize=9, color='magenta', fontweight='bold')

plt.tight_layout()
plt.show()

print(f"Group 1 average: [{avg_group1_3d[0]:.2f}, {avg_group1_3d[1]:.2f}, {avg_group1_3d[2]:.2f}]")
print(f"Group 2 average: [{avg_group2_3d[0]:.2f}, {avg_group2_3d[1]:.2f}, {avg_group2_3d[2]:.2f}]")
print(f"Incoming query: [{incoming_query[0]:.2f}, {incoming_query[1]:.2f}, {incoming_query[2]:.2f}]")
```
