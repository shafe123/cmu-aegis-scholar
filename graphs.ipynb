```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values from 0 to 1
x = np.linspace(0, 1, 100)

# Calculate y = 1 - x
y = 1 - x

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.xlabel('Vector Distance')
plt.ylabel('Converted Score')
plt.title('Distance Conversion')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values from 0 to 1200
x = np.linspace(0, 1200, 1000)

# Calculate y = 1 / (1 + e^(-0.005*(x-400)))
y = 1 / (1 + np.exp(-0.005 * (x - 400)))

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.xlabel('Number of Citations')
plt.ylabel('Citation Score')
plt.title('Citation Score vs Number of Citations')
plt.xlim(0, 1200)
plt.ylim(0, 1)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values from 0 to 40
x = np.linspace(0, 40, 1000)

# Calculate y = (-tanh(0.1*x - 2) + 1) / 2
y = (-np.tanh(0.1 * x - 2) + 1) / 2

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.xlabel('Years')
plt.ylabel('Time Score')
plt.title('Time Score vs Years')
plt.xlim(0, 40)
plt.ylim(0, 1)
plt.grid(True, alpha=0.3)
plt.show()
```
